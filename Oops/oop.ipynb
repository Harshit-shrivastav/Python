```python
class student:
    # defining default variable that user wouldn't pass
    clg = "IITM"
    def __init__(self, name, marks):
        self.name = name
        self.mark = marks
    def welcome(self):
        print("welcome", self.name)
    def get_marks(self):
        return self.mark
    # if we don't need to use self argument then @staticmethod decorator is used.
    @staticmethod
    def hi():
        print("Hello world")

s1 = student("Karan", 88)
print(s1.name, s1.mark)
print(s1.clg)
s1.welcome()
print(s1.get_marks())
s1.hi()
```

# Practice Problem

Q1.
Create student class that takes name and marks of 3 subject is as arguments in constructors, then create a method to print the average

```python
class students:
    def __init__(self, name, sub1, sub2, sub3):
        self.name = name
        self.sub1 = sub1
        self.sub2 = sub2
        self.sub3 = sub3
    def avg(self):
        a = (self.sub1 + self.sub2 + self.sub3)/3
        return a
n = students("Harshit", 99, 97, 100)
print(n.avg())
```

## Inheritance in oops

```python

class Person():
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(self.name, self.age)


class Student(Person):
    def __init__(self, name, age, marks):
        super().__init__(name, age)
        self.marks = marks

    def display(self):
        super().display()
        print(self.marks)


class Employee(Person):
    def __init__(self, name, age, salary):
        super().__init__(name, age)
        self.salary = salary

    def display(self):
        super().display()
        print(self.salary)


s = Student('Harshit', 18, 100)
s.display()
e = Employee('Harshit', 25, 10000000)
e.display()
```
